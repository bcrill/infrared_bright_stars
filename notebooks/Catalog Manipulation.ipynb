# Examples for reading and manipulating catalogs

```python
# import modules

from astropy.table import Table
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
```

```python
indir = '/Volumes/Untitled/'
```

```python
Table_2Mass = Table.read( indir+"2MASS/bright2mass.fits")
Table_Tycho2 = Table.read( indir+'Tycho-2/Tycho-2.fits')
```

```python
Table_Tycho2['BT'].max()
```

## A histogram of Tycho VT magnitudes

```python
y, bin_edges = np.histogram(Table_Tycho2[Table_Tycho2['VT']<30]['VT'],bins=100)
x = 0.5 * (bin_edges[1:]+bin_edges[:-1])
dm = x[1]-x[0]
plt.plot(x,y/dm)
plt.semilogy()
plt.xlabel('Tycho-2 VT [mag]')
plt.ylabel('Counts / mag')
plt.title(f'Tycho-2 sources n={len(Table_Tycho2)}')
plt.show()
```

```python
## A histogram of 2MASS J-band magnitudes
```

```python
values = Table_2Mass['J'].to(u.ABmag).value
values = values[np.where(values<30)]
y, bin_edges = np.histogram(values,bins=100)
x = 0.5 * (bin_edges[1:]+bin_edges[:-1])
dm = x[1]-x[0]
plt.plot(x,y/dm)
plt.semilogy()
plt.xlabel('2MASS J [mag]')
plt.ylabel('Counts / mag')
plt.title(f'2MASS bright sources n={len(Table_2Mass)}')

plt.show()
```

```python
plt.plot(Table_2Mass["ra"][::1000], Table_2Mass["dec"][::1000],',')
```

```python
YBS = Table.read( indir+"BSCV5/BSCV5.fits")
```

```python
YBS
```

```python
values = YBS['Vmag']
values = values[np.where(values<30)]
y, bin_edges = np.histogram(values,bins=100)
x = 0.5 * (bin_edges[1:]+bin_edges[:-1])
dm = x[1]-x[0]
plt.plot(x,y/dm)
plt.semilogy()
plt.xlabel('Vmag [mag]')
plt.ylabel('Counts / mag')
plt.title(f'Yale Bright Star catalog n={len(YBS)}')

plt.show()
```

```python
T = YBS[YBS["GLON"]<360.]

plt.plot(T["GLON"],T["GLAT"],',')
plt.show()
```

```python
from astropy.coordinates import SkyCoord
```

```python
C_YBS = SkyCoord(T["GLON"]*u.deg,T["GLAT"]*u.deg,frame='galactic')
```

```python
C_2MASS = SkyCoord(Table_2Mass["ra"]*u.deg, Table_2Mass["dec"]*u.deg,frame='icrs')
```

```python
C_YBS
```

```python
C_2MASS
```

```python
idx, d2d, d3d = C_YBS.match_to_catalog_sky(C_2MASS)
```

```python
plt.plot(d2d)
plt.ylabel(f'd2d {d2d.unit} ')
plt.ylim(0,0.01)
plt.show()
```

```python
matched_2MASS = Table_2Mass[idx]
```
